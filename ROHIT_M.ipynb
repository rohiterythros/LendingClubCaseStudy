### Bivariate Analysis

```python
#Bivariate Analysis - Loan status vs Annual income

ls_vs_inc= loan.groupby(['annual_inc_categories', 'loan_status']).loan_status.count().unstack().fillna(0).reset_index()
ls_vs_inc['Total'] = ls_vs_inc['Charged Off'] + ls_vs_inc['Current'] + ls_vs_inc['Fully Paid'] 
ls_vs_inc['Chargedoff_Proportion'] = ls_vs_inc['Charged Off'] / ls_vs_inc['Total']
ls_vs_inc['Fully_paid_Proportion'] = ls_vs_inc['Fully Paid'] / ls_vs_inc['Total']
ls_vs_inc.sort_values('Chargedoff_Proportion', ascending=False)

# Observations:
# The maximum proportion of the number of applicants charged off is in the category of annual income 0 to 20000, minimum is 80000 - 6000000 
# The minimum proportion of the number of applicants fully paid is in the category of 0 to 20000, maximum is 80000 - 6000000.

# Therefore annual income is a good indicator of loan repayability in the future.
# The more the annual income, the more they are likely to pay it back.
```

```python
#Bivariate Analysis - Loan status vs Purpose

ls_vs_purpose = loan.groupby(['purpose', 'loan_status']).loan_status.count().unstack().fillna(0).reset_index()
ls_vs_purpose['Total'] = ls_vs_purpose['Charged Off'] + ls_vs_purpose['Current'] + ls_vs_purpose['Fully Paid'] 
ls_vs_purpose['Chargedoff_Proportion'] = ls_vs_purpose['Charged Off'] / ls_vs_purpose['Total']
ls_vs_purpose['Fully_Paid_Proportion'] = ls_vs_purpose['Fully Paid'] / ls_vs_purpose['Total']
ls_vs_purpose.sort_values('Chargedoff_Proportion', ascending=False)

# Observations:
# The maximum proportion of the number of applicants charged off is in the category of 'Small Business', minimum is During a major purchase, Car loans and Wedding loans. 
# The minimum proportion of the number of applicants fully paid is in the category of 'Small business', maximum is 'Major purchase' and 'Wedding'.

# Small businesses seem to have the highest risk, hence it is best to charge them higher rates of interest.
```

```python
#Bivariate Analysis - Loan status vs Loan amount

ls_vs_loan_amt= loan.groupby(['loan_amnt_categories', 'loan_status']).loan_status.count().unstack().fillna(0).reset_index()
ls_vs_loan_amt['Total'] = ls_vs_loan_amt['Charged Off'] + ls_vs_loan_amt['Current'] + ls_vs_loan_amt['Fully Paid'] 
ls_vs_loan_amt['Chargedoff_Proportion'] = ls_vs_loan_amt['Charged Off'] / ls_vs_loan_amt['Total']
ls_vs_loan_amt['Fully_paid_Proportion'] = ls_vs_loan_amt['Fully Paid'] / ls_vs_loan_amt['Total']
ls_vs_loan_amt.sort_values('Chargedoff_Proportion', ascending=False)

# Observations:
# The maximum proportion of the number of applicants charged off is in the category of 30000 - 350000, minimum is 5000 - 10000 
# The minimum proportion of the number of applicants fully paid is in the category of 30000 - 35000, maximum is 5000 - 10000

# According to a higher value of their fully paid proportion and a lower value of their charged off proportion, the loan amount ranges can be charged a lower or higher interest rate.

# It is interesting to note here that the most solvent group is in the neither too high nor too low range of loan amount
```

```python
#Bivariate Analysis - Loan status vs Grade

ls_vs_grade = loan.groupby(['grade', 'loan_status']).loan_status.count().unstack().fillna(0).reset_index()
ls_vs_grade['Total'] = ls_vs_grade['Charged Off'] + ls_vs_grade['Current'] + ls_vs_grade['Fully Paid'] 
ls_vs_grade['Chargedoff_Proportion'] = ls_vs_grade['Charged Off'] / ls_vs_grade['Total']
ls_vs_grade['Fully_Paid_Proportion'] = ls_vs_grade['Fully Paid'] / ls_vs_grade['Total']
ls_vs_grade.sort_values('Chargedoff_Proportion', ascending=False)

# Observations:
# The maximum proportion of the number of applicants charged off is in the category of 'G', minimum is 'A' 
# The minimum proportion of the number of applicants fully paid is in the category of 'G' and 'F', maximum is 'A'

# Here the grade is a very good indicator of the applicant's solvency.
# Higher Interest rates can be charged according to how bad their grades are.
# Also lending could be denied to aaplicants with a grade of F or lower.
```
